```python
from easynmt import EasyNMT
```

```python
model= EasyNMT('opus-mt')
```

```python
sentences=['Jesus is alive!',
           'Hands on project is our priority',
           'I do not have anything',
           'It is a rainy day',
           "'It is what it is!':Legend Belise once said."]
```

```python
translations_fr= model.translate(sentences, target_lang='fr')
translations_rw= model.translate(sentences, target_lang='rw')
translations_ch= model.translate(sentences, target_lang='zh')
translations_sw= model.translate(sentences, target_lang='sw')
```
